```python
x = 10
```

```python
print("Value of x: "+ str(x))
```

```python
y = "Hello"
```

```python
print("Value of y: "+y)
```

```python
print("Value of x: "+ str(x) + "\n" + "Value of y: "+ y)
```

```python
# Now we shall use Interpolation
print("Value of x: %s"%(x))
```

```python
print("Value of x:%s \nValue of y:%s "%(x,y))
```

```python
print("Value of x: {}".format(x))
```

```python
print("Value of x:{a1} \nValue of y:{a2}".format(a1=x,a2=y))
```

```python
print(f"Value of x: {x}")
```

```python
print(f"Value of x:{x} \nValue of y:{y}")
```
