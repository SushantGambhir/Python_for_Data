```python
l = [1,2,1,3,1]
print(l)
```

```python
l.count(1)
```

```python
l.count(6)
```

```python
l2 = l
print(l2)
l.append(5)
print(l2)
```

```python
l = [1,2,1,3,1]
l2 = l.copy()
print(l2)
l.append(5)
print(l2)
```
